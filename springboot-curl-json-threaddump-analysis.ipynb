```python
from requests import request
import pandas as pd
from IPython.display import display, HTML
```

```python
response=request(url='http://localhost:8081/actuator/threaddump', method='get',headers={"Accept": "application/json"})
data = pd.read_json(response.text,orient='threads')
df = pd.json_normalize(data['threads'])
display(HTML('<H3>Top 5 Rows from Threaddump - No particular order</H3>'))
display(df.head(5))
display(HTML('<H3>List state of particular thread names</H3>'))
df[df['threadName'].str.startswith('reactor')][['threadName','threadState','suspended']]
```

```python
stack = df[df['threadName']=='reactor-http-epoll-1']['stackTrace'].to_dict()
## if you see in the table above, the index for 'reactor-http-epoll-1' is 30, in case you are wondering !!!
dfs = pd.DataFrame(stack[30],columns=['level','classLoaderName','moduleName','moduleVersion'\
                                  ,'methodName','fileName','lineNumber','className','nativeMethod'])
display(HTML('<H3>StackTrace for a particular thread</H3>'))
display(dfs)
```
